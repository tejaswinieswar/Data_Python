# Question 8:
Find a word in nps_chat but not in wordlist, it is considered an unusual word. (e.g. Heyyyy, gr8, 4ever can be considered unusual words for this analysis)
Find out how many unusual words people speak in chat (notice: usernames should not be counted)


```python
import nltk,string
from nltk.corpus import nps_chat, words
```

```python
a=nps_chat.fileids()
w_nps=[]
unusual=[]
for i in a:#Looping through the files of nps_chat
    x=nltk.corpus.nps_chat.words(i)#Extracting the words from each file
    w_nps.extend(x)#Appending the extracted words to a list
y=nltk.corpus.words.words("en")+nltk.corpus.words.words("en-basic")#Combining files in words package
y1=[]
for i in y:#Iteration to convert all dictionary words to lower case
    y1.append(i.lower())#Appendning the lower case words to a list y1
w_nps=set(w_nps)#To get the distinct chat words
print("The distinct number of words in nps_chat is",len(w_nps))
```

```python
for i in w_nps:
    if i!='' and (i.lower() not in y1):#Convert the word to lower and check if it exists in the dictionary
        if(i.startswith("U") and i[1].isdigit()) or i.isdigit() or (i[0] in (set(string.punctuation))):#To check and omit the user names and plain numbers and any punctuation
            #print (i)
            continue
        else:
            unusual.append(i)
print("The number of unusual words in nps_Chat:",len(unusual))
print ("The unusual words are\n",unusual)
```
