```python
#Importing the necessary pkgs
import glob
import csv
import os
import math
from math import log
import matplotlib.pyplot as plt
from nltk import FreqDist
from nltk.corpus import gutenberg
```

```python
a=glob.glob('/Users/Sneha/nltk_data/corpora/gutenberg/austen-emma.txt')#to open the file using glob in the path where the file resides
#print(a)
b=nltk.corpus.gutenberg.words('austen-emma.txt')
#print('The length of the words in the following text file is:',len(b))
for i in a:#
    v=open(i,'r')
    b=v.read()
    #b=v.read()
p=b.lower() #Converting all the data from text file to lower case
#print(p)    
j=''.join(m for m in p if (m.isalpha() or m==" "))    #Removing numbers and punctuations
#print (j)
fd=FreqDist(j.split())   #Calculating frequency distribution of words 
s=fd.most_common()   #Picking all the words
#print(s)
sh_words=[]
x=[]
y=[]

def f():#Using def function to put word freq and count to the excel file
    csvfile=open('output_file.csv', 'a')
    title='word'+','+'count'+','+'freq'+'\n'
    csvfile.write(title)
    count=1
    for i in s:
        word,freq= i
        sh_words.append(word)
        print(i,count)
        txt=lambda w,c,f: str(w)+','+str(c)+','+str(f)+'\n'
        text1=txt(word,count,freq)
        csvfile.write(text1)
        count=count+1
        g=math.log(count)
        x.append(g)
        #print(g)
        h=math.log(freq)
        y.append(h)
        #print(h)
        #math.plot(g,h)
#csvfile.closed()
if os.path.isfile('/Users/Sneha/output_file.csv'):
    os.remove('/Users/Sneha/output_file.csv')
f()#Calling the def function to show the word freq and rank
plt.plot(x,y,'r.',x,y,'b-')#Plotting the log(freq) vs log(count)
plt.xlabel('log(count)')
plt.ylabel('log(freq)')
plt.show()

```
